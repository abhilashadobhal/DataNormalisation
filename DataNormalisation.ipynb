#data Normalisation 
#Methods are simple feature scaling,Min-Max,z-Score or Standard Score
#Simple Feature scaling   x(new)=x(old)/xmax   Result =0 to 1
#Min Max    x(nex)=(x(old)-x(min))/(x(max)-x(min))  result=0 to 1
# z-Score x(new)=(x(old) -mu)/sigma)        mu=Average of feature  and sigma - Standard deviation   Result =-3 to 3

```python
headers=["S.No","Even","Odd","Multiple of 5","Multiple of 3","Multiple of 4","Length"]
df.columns=headers
print(df)
```

```python
#Simple Feature scaling   x(new)=x(old)/xmax   Result =0 to 1
df["Length"]=df["Length"]/df["Length"].max()
df.head()
```

```python
#Min Max    x(new)=(x(old)-x(min))/(x(max)-x(min))  result=0 to 1
path ="C:/Users/Guggi/Desktop/data analysis 1.xlsx"
df =pd.read_excel(path)
print(df)
```

```python
#Min Max    x(new)=(x(old)-x(min))/(x(max)-x(min))  result=0 to 1
df["Length"]=(df["Length"]-df["Length"].min())/(df["Length"].max()-df["Length"].min())
print(df)
```

```python
# z-Score x(new)=(x(old) -mu)/sigma)        mu=Average of feature  and sigma - Standard deviation   Result =-3
df["Length"]=(df["Length"]-df["Length"].mean())/df["Length"].std()
df.head()
```
