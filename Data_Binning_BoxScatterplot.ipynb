```python
#one-Hot Encoding is a technique used in data preprocessing to convert
#categorical data (text labels) into numerical format so that machine 
#learning models can understand it.

Most ML algorithms cannot work directly with text like
#"Red" or "Male", so we convert categories into binary (0 and 1) columns.
```

```python
# One Hot Encoding
S No	Fuel	 S.No	Fuel	   Gasoline	Diesal
1	Gasoline	 1	  Gasoline	       1	0
2	Diesal	     2	  Diesal	       0	1
3	Diesal	     3	  Diesal	       0	1
4	Gasoline	 4	  Gasoline	       1	0
```

```python
path ="C:/Users/Guggi/Desktop/data analysis 1.xlsx"
df =pd.read_excel(path)
print(df)
```

```python
#Numeric to Catregorical
bins =np.linspace(min(df["Length"]),max(df["Length"]),4)
```

```python
group_names =["Low","Medium","High"]
```

```python
df["Length-binned"]=pd.cut(df["Length"],bins,labels=group_names,include_lowest= True)
```
